```python
text ="""ربما كانت أحد أهم التطورات التي قامت بها الرياضيات العربية التي بدأت في هذا الوقت بعمل الخوارزمي وهي بدايات الجبر، ومن المهم فهم كيف كانت هذه
الفكرة الجديدة مهمة، فقد كانت خطوة نورية بعيدا عن المفهوم اليوناني للرياضيات التي هي في جوهرها هندسة، الجبر کان نظرية موحدة تتيح الأعداد الكسرية
والأعداد اللا كسرية، والمقادير الهندسية وغيرها، أن تتعامل على أنها أجسام جبرية، وأعطت الرياضيات ككل مسارا جديدا للتطور بمفهوم أوسع بكثير من الذي كان
موجودا من قبل، وقم وسيلة للتنمية في هذا الموضوع مستقبلا. وجانب آخر مهم لإدخال أفكار الجبر وهو أنه سمح بتطبيق الرياضيات على نفسها بطريقة لم تحدث
من قبل """
print(text)
```

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

# Word token

```python
from nltk import pos_tag
from nltk import RegexpParser
from nltk.tokenize import sent_tokenize

tokens = nltk.word_tokenize(text)
print("After Split:",tokens)
```

# Sentance token

```python
sent=sent_tokenize(text)
print("After sentence token:",sent)
```

# POS

```python
tokens_tag = pos_tag(tokens,lang='ae')
print("After Token:",tokens_tag)
```

# Use Case of Chunking


```python
grammar = "NP: {<DT>+<JJ>+<NN>+}"
cp =nltk.RegexpParser(grammar)
result = cp.parse(tokens_tag)
print(result)
```

# Stemming

```python
import nltk
stemming=[]
for i in tokens: 
  stemm = nltk.stem.PorterStemmer()
  stemming.append(stemm.stem(i))
print(stemming)
```

# Lemmatization

```python
Lemmatization=[]
for i in tokens: 
  lemma = nltk.wordnet.WordNetLemmatizer()
  a=lemma.lemmatize(i)
  Lemmatization.append(a)
print(Lemmatization)
```
